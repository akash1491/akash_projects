```python
## randomly fill vehicle names based on above list
from numpy import random

rn.seed(0)
df['vech_name']= rn.choice(Vechiles_list,size=len(df))
```

```python
## let's take TN for example and fill the vechile_id
## 88
## let's fill the slot with 6 digt number, where 1st 2 digits are between 0 and 88, last 4 are between 0000 and 9999

rn.seed(0)

## RTO code

rto = []
for i in range(0,len(df)):
    n = random.randint(1,88)
    rto.append(n)

## Vechile Number
    
vechile_digits = []
for i in range(0,len(df)):
    n = random.randint(1000,9999)
    vechile_digits.append(n)    
#print(randomlist)
```

```python
## once the car randomization we will fill the car number

## car number = TN + any 2 digits for RTO + any 4 digits for vehicle number
for i in range(0,len(df)):
    
    df['vech_no'].loc[i]='TN' + str(rto[i]) +'-' +str(vechile_digits[i])
```

```python
## state we will hardcode it as TN for now

df['State']='TN'
```

```python
## next step is to setup date ranges
## here we will take maxima and minima of and setup a randomization to pick values from this list
# initializing dates ranges
rn.seed(0)
start,end = date(2010, 1, 1), date(2023, 7, 1)
 
# printing dates
print("The original range : " + str(start) + " " + str(end))
 
# initializing K
K = len(df)
 
random_dates = [start]
 
# loop to get each date till end date
while start != end:
    start += timedelta(days=1)
    random_dates.append(start)
 
  # random K dates from pack
result = rn.choice(random_dates,size=len(df))
# printing
#print("K random dates in range : " + str(res))

df['Date of Event']= pd.to_datetime(result)
```
